```python
import geopandas as gpd
# Read the GeoJSON file
gdf = gpd.read_file('20_min_krs_car.json')
gdf.explore(
    column='color',  # Color based on the 'value' column
    cmap='viridis',  # Color map
    tooltip=['label', 'area', 'total_pop'],  # Show these columns on hover
    popup=True,      # Enable popup for more details
    style_kwds={'fillOpacity': 0.3}
)
```

```python
import pyarrow
import subprocess


# Convert the GeoDataFrame to GeoParquet
output_file = '20_min_krs_car.parquet'
gdf.to_parquet(output_file, schema_version="1.1.0", compression='snappy', write_covering_bbox=True)

print(f"GeoDataFrame successfully converted and saved to {output_file}")

# Optional: Verify the conversion by reading it back
verify = gpd.read_parquet(output_file)
print(f"Verification: Read {len(verify)} rows from the parquet file")
```
